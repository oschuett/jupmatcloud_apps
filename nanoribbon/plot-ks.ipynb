```python
import argparse, glob,re
import numpy as np
import matplotlib.pyplot as pl
import matplotlib.image as im
import sys
from matplotlib import rc
from matplotlib.backends.backend_pdf import PdfPages
from aiida.orm import load_node
```

```python
# # Define command line parser
# parser = argparse.ArgumentParser(
#     description='Plots list of Kohn-Sham orbitals into ks_orbitals.png,'
#                 'ks_orbitals.pdf')
# parser.add_argument('--version', action='version', version='%(prog)s 06.10.2016')
# parser.add_argument(
#     'node',
#     metavar='nodenumber', 
#     type=int,
#     help='Calculation node')
# parser.add_argument(
#     '--band',
#     type=int,
#     nargs='+',
#     help='Number of bands',
#     default=[])
# parser.add_argument(
#     '--kpoint',
#     type=int,
#     nargs='+',
#     help='Number of kpoints',
#     default=[])
# parser.add_argument(
#     '--dz',
#     type=float,
#     help='Height',
#     default=8.5)
# args = parser.parse_args()


class Foo():
    pass


args = Foo()
args.node = 3729
args.band = []
args.kpoint = []
args.dz = 8.5

# Get the list
node=load_node(args.node)
dz=args.dz
path=node.out.retrieved.get_abs_path()+'/path'
pattern=str(path + '/KS*.png')
#print pattern
allpngs=glob.glob(pattern)

bands=args.band
kpoints=args.kpoint
if len(bands) == 0:
    for f in allpngs:
        b=re.findall(r'_B-\d+',f)[0][3:]
        bands.append(b)
if len(kpoints) == 0:
    for f in allpngs:
        k=re.findall(r'_K-\d+',f)[0][3:]
        kpoints.append(k)

bands = list(set(bands))
kpoints=list(set(kpoints))
bands.sort(key=int,reverse=True)
kpoints.sort(key=int)

nrows=len(bands)
ncolumns=len(kpoints)
print ncolumns,nrows
print node,bands,kpoints

#exit()
# Prepare the figure
matrixname = 'ks_orbitals'
pp = PdfPages(str(matrixname + '.pdf'))
fig = pl.figure() 
fig.suptitle("Rows 1 to 4 are KS orbitals of Lumo+1 to  HOMO-1 bands at k points from G to X (columns 1 to 10)", fontsize=8)

font = { 'size'   : 6}
rc('font', **font)

# If number of columns is not set, make 1 column

i=0
for b in bands:
    for k in kpoints:
        i = i + 1
        myplot = pl.subplot(nrows,ncolumns, i)
        myplot.xaxis.set_visible(False)
        myplot.yaxis.set_visible(False)
#       filename="KS_B-{:03d}_K-{:02d}.dz{}.png".format(b,k,dz)
        filename="KS_B-{}_K-{}.dz{}.png".format(b,k,dz)
#       filename_t="B{}K{}dz{}".format(int(b),int(k),dz)
        filename_t="B{}_K{}".format(int(b),int(k))
        print filename
        img = im.imread(path+'/'+filename)
        imgplot = pl.imshow(img)
#       t = "%(fn)s" % dict(fn=filename)
        t = "%(fn)s" % dict(fn=filename_t)
        pl.title(t)
    
delta=0.05
pl.subplots_adjust(wspace=0,hspace=delta*8, bottom=delta,top=1-delta,right=1-delta,left=delta)

# To increase resolution of images in pdf, you need to increase the figure size.
# Note: You will need to adjust the font size.
xin = 10
fig.set_size_inches(xin, xin)

# For png output, the resolution can be controlled by the dpi setting.
pl.savefig(str(matrixname + '.png'), dpi=450)
#pl.show()

pp.savefig()
pp.close()
```
