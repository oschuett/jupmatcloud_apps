```python
import argparse, glob,re
import numpy as np
import matplotlib.pyplot as pl
import matplotlib.image as im
import sys
from matplotlib import rc
from matplotlib.backends.backend_pdf import PdfPages
from aiida.orm import load_node

# # Define command line parser
# parser = argparse.ArgumentParser(
#     description='Plots list of Kohn-Sham orbitals into ks_orbitals.png,'
#                 'ks_orbitals.pdf')
# parser.add_argument('--version', action='version', version='%(prog)s 06.10.2016')
# parser.add_argument(
#     'node',
#     metavar='nodenumber', 
#     type=int,
#     help='Calculation node')
# parser.add_argument(
#     '--dz',
#     type=float,
#     help='Height',
#     default=2.0)
# #parser.add_argument(
# #    '--de',
# #    type=float,
# #    help='Height',
# #    default=0.05)


# args = parser.parse_args()

class Foo():
    pass

args = Foo()
args.node = 3734
args.dz = 2.0
args.de = 0.05

# Get the list
node=load_node(args.node)
dz=args.dz
print dz
path=node.out.retrieved.get_abs_path()+'/path'
pattern=str(path + '/sts_*'+str(dz)+'*.png')
#print pattern
allpngs=glob.glob(pattern)

energies=[]
filenames={}
nene=len(allpngs)
semin=re.findall(r'[-+]?\d*\.\d+',allpngs[0])[0]
semax=re.findall(r'[-+]?\d*\.\d+',allpngs[0])[1]
semin=float(semin)
semax=float(semax)
de=(semax-semin) / (float(nene))
print semin, semax, nene,de

i=0
for f in allpngs:
    e=semin+de*i
    filenames [str("{0:.2f}".format(round(e,2)))] = f
    i+=1
    energies.append(round(e,2))

energies = list(set(energies))
energies.sort(key=float)
#energies.sort(key=str)
print "list of energies"
print energies
print"give me emin emax "
#emin,emax = map(float,sys.stdin.readline().split())
#emin,emax = map(float,sys.stdin.readline().split())
emin, emax = -6, -5
print emin,emax



# Prepare the figure
matrixname = 'STS'
pp = PdfPages(str(matrixname + '.pdf'))
fig = pl.figure() 
font = { 'size'   : 6}
rc('font', **font)

# If number of columns is not set, make 1 column

ncolumns=6
nrows=3
nenergies=0
todofiles=[]

for e in energies:
    #lae=float(e[2:])   
    if e >= emin and e <=emax :
        nenergies+=1
        todofiles.append(filenames[str("{0:.2f}".format(round(e,2)))])

nrows = nenergies/ncolumns +1

i=0
for ilf in todofiles:
        i = i + 1
        myplot = pl.subplot(nrows,ncolumns, i)
        myplot.xaxis.set_visible(False)
        myplot.yaxis.set_visible(False)
        filename=ilf
        filename_t=" "
        print filename
#       img = im.imread(path+'/'+filename)
        img = im.imread(filename)
        imgplot = pl.imshow(img)
        t = "%(fn)s" % dict(fn=filename_t)
        pl.title(t)
    
delta=0.05
pl.subplots_adjust(wspace=0,hspace=delta*8, bottom=delta,top=1-delta,right=1-delta,left=delta)

# To increase resolution of images in pdf, you need to increase the figure size.
# Note: You will need to adjust the font size.
xin = 10
fig.set_size_inches(xin, xin)

# For png output, the resolution can be controlled by the dpi setting.
pl.savefig(str(matrixname + '.png'), dpi=450)
#pl.show()

pp.savefig()
pp.close()
```
